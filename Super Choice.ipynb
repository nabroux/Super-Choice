```python
import random, datetime

rankList = []
damnDict = {}

def picking(p0, p1):
    print(p0,'or', p1)
    while True:
        winnerIndex = input('who wins? 0 or 1 or damn: ')
        if winnerIndex == '0':
            winner = p0
            loser = p1
            break
        elif winnerIndex == '1':
            winner = p1
            loser = p0
            break
        elif winnerIndex == 'damn':
            damnReason = input('Tell us why: ')
            damnDict[p0 +' v.s '+ p1] = damnReason
            print('Ok.. but you still have to choose.\n')
            print(p0,'or', p1)
        else:
            print(' 0 or 1 or damn only! ')
            
    return (winner, loser)

def findOp(player, rankList):
    mini_diff = 18
    if gender == 'b':
        for n in rankList:
            op_index = boys.index(n)
            pl_index = boys.index(player)
            diff = abs(op_index-pl_index)
            if diff < mini_diff:
                mini_diff = diff
                newOp = n
    elif gender == 'g':
        for n in rankList:
            op_index = girls.index(n)
            pl_index = girls.index(player)
            diff = abs(op_index-pl_index)
            if diff < mini_diff:
                mini_diff = diff
                newOp = n
    return newOp
                
girls = '''鄭捷云
趙秉之
蔡宜臻
彭子瑜
莊景雲
陳廷瑜
陳倩文
蕭郁儒
李亮慧
彭鈺雯
姚佳妤
張躍馨
詹雅筑
許明敏
孫和湉
李睿訢
張家芸
林雨萱'''.split('\n')

boys = '''翁典
劉安哲
徐周豪
陳品嘉
張鎧
梁瑞允
戴士捷
彭詠惟
黃詣凱
陳尚謙
鮑威宇
莫力全
張哲銓
謝杰修
蘇祐陞
余柏霖
許雲輔
林承叡'''.split('\n')

name = input('請輸入您的姓名: ')

while True:
    gender = input('您要排名男生還是女生(請輸入 b or g): ')
    if gender == 'b':
        newList = boys[:]
        break
    elif gender == 'g':
        newList = girls[:]
        break
    else:
        print('b or g only!')

#first round
players = random.sample(newList, 2)
newList.remove(players[0])
newList.remove(players[1])
winner, loser = picking(players[0], players[1])
rankList.insert(0, winner)
rankList.append(loser)

#the rest
while newList:
    player = random.choice(newList)
    newList.remove(player)
    newOp = findOp(player, rankList)
    opIndex = rankList.index(newOp)
    winner, loser = picking(player, newOp)
    if winner == player:
        if opIndex != 0:
            while True:
                opIndex -= 1
                winner, loser = picking(player, rankList[opIndex])
                if loser == player:
                    rankList.insert(opIndex + 1, player)
                    break
                elif winner == player and opIndex == 0:
                    rankList.insert(0, player)
                    break
        elif opIndex == 0:
            rankList.insert(0, player)

    elif loser == player:
        if opIndex != len(rankList)-1:
            while True:
                opIndex += 1
                winner, loser = picking(player, rankList[opIndex])
                if winner == player:
                    rankList.insert(opIndex, player)
                    break
                elif loser == player and opIndex == len(rankList)-1:
                    rankList.append(player)
                    break
        elif opIndex == len(rankList)-1:
            rankList.append(player)

print('排名列表：\n',rankList)
print('筆記列表：\n',damnDict)

now = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
data = '\n' + (','.join([name,gender,str(rankList),str(damnDict),now]))
with open('history.csv', 'a', encoding='utf-8') as f:
    f.write(data)
```
